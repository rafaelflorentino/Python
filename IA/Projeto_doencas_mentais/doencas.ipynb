# Análise Temporal da Prevalência de Transtornos Mentais (Média Anual Global)

```python
import matplotlib.pyplot as plt

# Carregar o dataset
url = "../datasets/mental-illnesses-prevalence.csv"
dados = pd.read_csv(url, sep=",")

# Filtrar os dados por um país específico (caso queira, por exemplo, 'Afghanistan')
# país = "Afghanistan"
# dados_pais = dados[dados['Entity'] == país]

# Caso queira a média anual global (considerando todos os países), use o seguinte:
dados_ano = dados.groupby('Year').mean(numeric_only=True).reset_index()

# Plotando os dados
plt.figure(figsize=(12, 6))
plt.plot(dados_ano['Year'], dados_ano['Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized'], label="Esquizofrenia")
plt.plot(dados_ano['Year'], dados_ano['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized'], label="Depressão")
plt.plot(dados_ano['Year'], dados_ano['Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized'], label="Ansiedade")
plt.plot(dados_ano['Year'], dados_ano['Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized'], label="Bipolar")
plt.plot(dados_ano['Year'], dados_ano['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'], label="Transtornos Alimentares")

# Personalizando o gráfico
plt.title("Análise Temporal da Prevalência de Transtornos Mentais (Média Anual Global)")
plt.xlabel("Ano")
plt.ylabel("Prevalência (%) da População")
plt.legend()
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

# Dados do Brasil

```python
# Carregar o dataset
url = "../datasets/mental-illnesses-prevalence.csv"
dados = pd.read_csv(url, sep=",")

# Filtrar os dados para o Brasil
dados_brasil = dados[dados['Entity'] == 'Brazil']

# Definir o ano como eixo X
anos = dados_brasil['Year']

# Criar o gráfico para cada transtorno
plt.figure(figsize=(12, 8))

plt.plot(anos, dados_brasil['Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized'], label='Esquizofrenia')
plt.plot(anos, dados_brasil['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized'], label='Depressão')
plt.plot(anos, dados_brasil['Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized'], label='Ansiedade')
plt.plot(anos, dados_brasil['Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized'], label='Transtorno Bipolar')
plt.plot(anos, dados_brasil['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'], label='Transtornos Alimentares')

# Personalizar o gráfico
plt.xlabel('Ano')
plt.ylabel('Prevalência (%) da População')
plt.title('Análise Temporal da Prevalência de Transtornos Mentais no Brasil')
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
```

# Comparação entre  Paises

```python
# Carregar o dataset
url = "../datasets/mental-illnesses-prevalence.csv"
dados = pd.read_csv(url, sep=",")

# Selecionar 10 países para comparação
paises = ['Brazil', 'Argentina', 'Albania', 'Canada', 'Vietnam', 'Venezuela', 'Germany', 'India', 'Italy', 'Australia']
dados_selecionados = dados[dados['Entity'].isin(paises)]

# Filtrar os dados para anos específicos, se necessário (por exemplo, 2010-2019)
dados_selecionados = dados_selecionados[dados_selecionados['Year'] >= 2010]

# Pivotar os dados para facilitar a plotagem
dados_pivotados = dados_selecionados.pivot_table(index='Year', columns='Entity', 
                                                   values='Depressive disorders (share of population) - Sex: Both - Age: Age-standardized')

# Plotar o gráfico
plt.figure(figsize=(14, 8))
for pais in paises:
    plt.plot(dados_pivotados.index, dados_pivotados[pais], marker='o', label=pais)

# Personalizar o gráfico
plt.title('Comparação da Prevalência de Transtornos Depressivos entre 10 Países (2010-2019)')
plt.xlabel('Ano')
plt.ylabel('Prevalência (%) da População')
plt.xticks(dados_pivotados.index)  # Mostrar todos os anos
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
```

# Correlação entre trasntornos Mentais

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
url = "../datasets/mental-illnesses-prevalence.csv"
dados = pd.read_csv(url, sep=",")

# Selecionar dados para um país específico, por exemplo, Brasil
dados_brasil = dados[dados['Entity'] == 'Brazil']

# Selecionar apenas as colunas de interesse
colunas_interesse = [
    'Year',
    'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized',
    'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized',
    'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized',
    'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized',
    'Eating disorders (share of population) - Sex: Both - Age: Age-standardized'
]

# Filtrar o dataframe para manter apenas as colunas de interesse
dados_brasil_filtered = dados_brasil[colunas_interesse]

# Calcular a matriz de correlação
correlacao = dados_brasil_filtered.corr()

# Configurar o tamanho da figura
plt.figure(figsize=(10, 6))

# Criar um mapa de calor da matriz de correlação
sns.heatmap(correlacao, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True, 
            xticklabels=correlacao.columns, yticklabels=correlacao.columns)

# Adicionar título
plt.title('Correlação entre Transtornos Mentais no Brasil (Age-standardized)')

# Exibir o gráfico
plt.show()
```
