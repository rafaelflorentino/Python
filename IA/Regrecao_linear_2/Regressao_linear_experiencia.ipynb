Regressão linear = encontrar a relação entre duas variaveis uma independente(entrada) e uma dependente(Saida), gera uma linha reta que aproxima os potos, preve um valor de saida baseado no valor de entrada.

# Qual a diferença entre regressão linear e regressão logistica?

1. Enquanto a linear prever um valor, a lógistica preve uma probabilidade.
2. A probabilidade esta entre 0 e 1 (variável categorica)
3. A aplicação de lógistica frequentemente é binária.
4. Utiliza-se variáveis númericas contínuas na regressão linear, na lógistica também, em relação as variáveis independentes.

# Preparar os dados e as classes(rótulos)

```python
# selecionar linhas e colunas com iloc
# : significa que você está selecionando todas as linhas
# :-1 significa que você está selecionando todas as colunas, exceto a última.
X = df.iloc[:,:-1].values
X
```

```python
# Classes (rótulos)

y = df.iloc[:,-1].values
y
```

# Dividir em treino e teste

image.png

 - X1, X2, X3 - dados de treino
 - y rotulos(classes) de teino

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split - 4 valores
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state=0)
```

```python
# 67% dados de treino
X_train
```

```python
# 33% de dados de teste
X_test
```

```python
# 67% salários ( oque eu prever) - y (rótulos ou classes)
y_train
```

```python
# 33% salários ( o que eu quero prever) - y (rótulos ou classes)
y_test
```

# Aplicando regressão linear

```python
from sklearn.linear_model import LinearRegression
```

```python
# Criar uma nova instância do modelo de regressão linear
regressor = LinearRegression()
```

```python
# Treinar o modelo de regressão linear
regressor.fit(X_train, y_train)
```

# Aplicando testes

```python
# passando o 33.33333%
y_pred = regressor.predict(X_test)
```
