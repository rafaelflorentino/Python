```python
dados = {
    'X': [1,2,100]
}

dados = pd.DataFrame(dados)
```

```python
# Média, os valores da média são muito sensíveis a valores muito altos ou muito baixos, pode gerar discrepância nos dados.
# É bom remover os outliers ou usar outra métrica se for o caso.
dados.mean()
```

```python
# Mostrando a contagem de registros
dados.count()
```

```python
# Mostra a mediana
# A mediana não muda muito com valores muito altos e muito baixos, pois leva em conta a posição dos valores
# Ajuda a ver a distribuição dos dados
# Números precisam estar ordenados
dados.median()
```

```python
# Ver o desvio padrão
# Quanto os dados estão variando em relação a média
dados.std()
```

```python
# Mostra todos esses dados estatísticos
# 50% e a mediana
# 25% e 75% são os quartis, medidas de separatrizes (ver onde os dados estão concetrados)
dados.describe()
```

```python
# Veriificar se alguem viajou com esposa ou com filhos
# argumento logico | (ou)
base[(base.Parch > 1) | (base.SibSp > 1)].head()
```

```python
# Veriificar se alguem viajou com esposa e com filhos usando .loc
# com .loc posso subistituir valores
# argumento lógico & (e)
base.loc[(base.Parch > 1) & (base.SibSp > 1)].head()
```

```python
# Filtrar colunas
base.loc[(base.Parch > 1 ) | (base.SibSp > 1), ['Name', 'PassengerId', 'Parch', 'SibSp']].head()
# Filtrando dados passando condições
```

```python
# Buscando linhas usando indices, e iloc
base.iloc[30:40]
```

```python
# Buscando as as linhas de 20 a 40 e as colunas de 3 a 6
base.iloc[30:40, 3:6]
```
