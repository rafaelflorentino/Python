```python
# Url do dataset online
url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
```

```python
import pandas as pd
```

```python
# Carregar o Dataset
df = pd.read_csv(url)
```

```python
# Mostrar os dois primeiros itens
df.head(2)
```

```python
# Mostrar as variedades de iris
df['variety'].value_counts()
```

```python
# Alterar os nomes das colunas, tirar ponto e colocar underline
# Aplicação do List Comprehension
# [] - Lista
# string --> s
# Trocar p '.' por '_' , boa prática para não nugar algumas funções 
df.columns = [s.replace('.','_') for s in df.columns]
df.head(5)
```

```python
# Copia independente (copy() não altera as alterações originais) 
df_setosa = df.query('variety == "Setosa" ').copy()
df_virginica = df.query('variety == "Virginica" ').copy()
df_versicolor = df.query('variety == "Versicolor" ').copy()

print(df_setosa)
print(df_virginica)
print(df_versicolor)

# criar uma lista dos nomes das colunas de medidias
medidas = ['sepal_length','sepal_width','petal_length', 'petal_width']
```
